## Exo 1 (9 points)

Soit à résoudre le problème suivant ($t_f \gt 0$ est fixé) :

$$ \int_0^{t_f} u^2(t)\,\mathrm{d}t \to \min $$

avec pour dynamique

$$ \ddot{q}(t)=u(t),\quad u(t) \in \mathbf{R}, $$
$$ q(0)=\dot{q}(0)=0,\quad q(t_f)=1. $$

(Noter que $\dot{q}(t_f)$ est libre.)

### 1.1
On pose $x(t)=(q(t),\dot{q}(t))$, mettre la dynamique sous la forme $\dot{x}(t)=f(x(t),u(t))$ avec $f$ une fonction que l'on précisera.

**Réponse.** $f(x,u)=(x_2,u)$

### 1.2
Écrire le hamiltonien du problème.

**Réponse.** $H(x,p,u) = p^0|u|^2+p_1 x_2+p_2 u$

### 1.3
Écrire le système adjoint.

**Réponse.** $\dot{p}_1 = 0$, $\dot{p}_2 = -p_1$

### 1.4
Montrer que $p^0=0$ est impossible. (On posera par conséquent $p^0=-1$ dans la suite de l'exercice.)

**Réponse.** Si $p^0=0$, la condition de maximisation implique $p_2$ nul p.p. Donc $p_1=-\dot{p}_2$ est aussi nul, ce qui conduit à $(p^0,p)=(0,0)$ (impossible).

### 1.5
Appliquer la condition de maximisation du PMP.

**Réponse.** $u=p_2/2$

### 1.6
Appliquer les conditions de transversalité.

**Réponse.** $p_2(t_f)=0$

### 1.7
En déduire l'état adjoint $p_2$.

**Réponse.** Comme $p_2$ est affine, vu sa valeur finale, il existe $a \in \mathbf{R}$ tel que $p_2(t)=a(t-t_f)$.

### 1.8
En déduire finalement le contrôle optimal.

**Réponse.** En intégrant $\ddot{q}=u=(a/2)(t-t_f)$ on détermine $a$ pour que $q(t_f)=1$, et on trouve
$u(t)=3(t_f-t)/t_f^3$.

### 1.9
Compléter le code de tir simple ci-dessous.

**Réponse.** Voir code ci-dessous (on remplace $u$ par $p_2/2$ pour avoir le hamiltonien maximisé).

```julia
t0 = 0.
tf = 3.5
x0 = [ 0., 0. ]

# Hamiltonian
function h(t, x, p)
    r = p[2]^2/4. + p[1]*x[2] # **** À COMPLÉTER ****
    return r
end

# Makes flow from Hamiltonian
f = Flow(h)

# Shooting function
function shoot(p0)
    xf, pf = f(t0, x0, p0, tf) # **** À COMPLÉTER ****
    s = [ xf[1]-1.0, pf[2] ] # **** À COMPLÉTER ****
    return s
end
```

## Exo 2 (8 points)

Soit à résoudre le problème de temps minimum de dynamique

$$ \dot{x}_1(t)=u_1(t),\quad \dot{x}_2(t)=u_2(t),\quad u_1^2(t)+u_2^2(t) \leq 1, $$

et de conditions aux deux bouts $x(0)=(0,0)$ et $x_1(t_f)\cos\theta + x_2(t_f)\sin\theta = \ell$ où $\theta \in \mathbf{R}$ et $\ell \gt 0$ sont fixés.

### 2.1
Interpréter géométriquement le problème.

**Réponse.** On cherche le plus court chemin de $(0,0)$ jusqu'à la droite $x_1\cos\theta + x_2\sin\theta = \ell$.

### 2.2
Écrire le hamiltonien du problème.

**Réponse.** $H(x,p,u) = p^0+p_1 u_1+p_2 u_2$

### 2.3
Écrire le système adjoint.

**Réponse.** $\dot{p}_1=0$, $\dot{p}_2=0$

### 2.4
Montrer que l'état adjoint ne peut pas s'annuler.

**Réponse.** Si $p$ (constant d'après ce qui précède) s'annule en un point, il est nul partout. Alors $0=H=p^0$ (car $t_f$ est libre), donc $(p^0,p)=(0,0)$ (impossible).

### 2.5
Appliquer la condition de maximisation du PMP.

**Réponse.** Comme $p$ ne s'annule pas, $u=p/|p|=(p_1,p_2)/\sqrt{p_1^2+p_2^2}$.

### 2.6
Appliquer les conditions de transversalité.

**Réponse.** L'espace tangent à la droite (affine) est la droite (vectorielle) $x_1\cos\theta + x_2\sin\theta = 0$, à laquelle $p(t_f)$ doit être orthogonal. C'est donc que $p(t_f)$ est colinéaire à $(\cos\theta,\sin\theta)$.

### 2.7
En déduire la trajectoire optimale.

**Réponse.** Comme $u$ est colinéaire à $p$ et de norme $1$, on doit avoir $u=\pm(\cos\theta,\sin\theta)$ : pour atteindre la cible, il faut choisir $u=+(\cos\theta,\sin\theta)$.

### 2.8
En déduire le temps minimal.

**Réponse.** $t_f=\ell$ (ce qui correspond bien à la projection orthogonale de $(0,0)$ sur la droite cible).

### 3.3
On part de $(x(0),y(0),\theta(0))=(0,0,\theta_0)$ et on applique le contrôle $u(t)=1$ pour $t \in [0,t_1]$. Donner l'expression de $(x(t_1),y(t_1),\theta(t_1)))$.

**Réponse.** On pose $z(t)=x(t)+iy(t)$ et on a $z(t_1)=wt_1+(1/i)(e^{i\theta(t_1)}-e^{i\theta_0})$ et $\theta(t_1)=\theta_0+t_1$.

```julia
# Constraints 
@constraints(sys, begin
    x[1, 1] == x0
    y[1, 1] == y0
    θ[1, 1] == θ0
    x[2, 1] == x[1, P] # continuité arc 1->2
    y[2, 1] == y[1, P] # continuité arc 1->2 
    θ[2, 1] == θ[1, P] # continuité arc 1->2
    x[3, 1] == x[2, P] # continuité arc 2->3
    y[3, 1] == y[2, P] # continuité arc 2->3
    θ[3, 1] == θ[2, P] # continuité arc 2->3
    x[3, P] == xf
    y[3, P] == yf
    θ[3, P] == θf
    end)

# Dynamics: Crank-Nicolson scheme
for j in 1 : P-1
    @NLconstraints(sys, begin
        # x' = w + cos(theta)
        x[1, j+1] == x[1, j] + 0.5 * τ[1]*Δt * ( w + cos(θ[1, j]) + w + cos(θ[1, j+1]) )
        x[2, j+1] == x[2, j] + 0.5 * τ[2]*Δt * ( w + cos(θ[2, j]) + w + cos(θ[2, j+1]) )
        x[3, j+1] == x[3, j] + 0.5 * τ[3]*Δt * ( w + cos(θ[3, j]) + w + cos(θ[3, j+1]) )
        # y' = sin(theta) 
        y[1, j+1] == y[1, j] + 0.5 * τ[1]*Δt * ( sin(θ[1, j]) + sin(θ[1, j+1]) )
        y[2, j+1] == y[2, j] + 0.5 * τ[2]*Δt * ( sin(θ[2, j]) + sin(θ[2, j+1]) )
        y[3, j+1] == y[3, j] + 0.5 * τ[3]*Δt * ( sin(θ[3, j]) + sin(θ[3, j+1]) )
        # theta' = u
        θ[1, j+1] == θ[1, j] +       τ[1]*Δt * u[1]
        θ[2, j+1] == θ[2, j] +       τ[2]*Δt * u[2]
        θ[3, j+1] == θ[3, j] +       τ[3]*Δt * u[3]
    end)
end
```
